````python
# criando dataframe com as informações do dicionário
list_orgaos_iacm = []
for ano, fonte in dict_cumprimento_metas.items():
    for orgao, iacm in fonte.get('IACM').items():
        list_orgao_iacm = []
        list_orgao_iacm.append(ano)
        list_orgao_iacm.append(orgao)
        list_orgao_iacm.append(iacm)
        list_orgaos_iacm.append(list_orgao_iacm)

df_iacm = pd.DataFrame(list_orgaos_iacm, columns=['ANO','ORGAO','IACM'])
````

````python
# agrupando os valores da média do IACM por ano criando a tabela
df_iacm.groupby(['ANO']).agg(MEDIA=('IACM','mean')).to_html('tabela_A2.html')
````

````python
df_media_iacm = df_iacm.groupby(['ORGAO']).agg(MEDIA=('IACM','mean'))
````

````python
# alterando a disposição das informações dos anos em colunas
df_iacm_anos = df_iacm.pivot(index="ORGAO", columns='ANO', values="IACM")
````

````python
# junta as informações ano a ano com a média de 2021-2024 e salva em arquivo csv
pd.merge(df_iacm_anos, df_media_iacm, left_index=True, right_index=True).sort_values(
    by=['MEDIA']).to_csv('iacm_orgaos_anos_media.csv')
````

````python
# junta as informações ano a ano com a média de 2021-2024 e salva em arquivo csv
pd.merge(df_iacm_anos, df_media_iacm, left_index=True, right_index=True).sort_values(
    by=['MEDIA']).fillna('').to_html('tabela_A1.html')
````
